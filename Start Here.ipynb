# Bluesky Software Tutorials

## Where to start

For an introduction aimed at total beginners, start with [Hello Python and Jupyter](./tutorials/Hello%20Python%20and%20Jupyter.ipynb). To dive into data acquisition basics, see [Hello Bluesky](./tutorials/Hello%20Bluesky.ipynb).

From there, explore [the titles available](./tutorials/). Most can be done in any order. If any prior knowledge is assumed, recommended prerequisites will be linked at the top.
